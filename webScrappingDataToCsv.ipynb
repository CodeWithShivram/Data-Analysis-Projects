```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
import csv

# Step 1: Fetch the page
my_url = "https://www.ebay.com/sch/i.html?_nkw=laptops"
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()

# Step 2: Parse the page
page_soup = soup(page_html, "html.parser")
containers = page_soup.findAll("div", {"class": "s-item__info"})

# Step 3: Create a CSV file to write data
filename = "products.csv"
with open(filename, "w", newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    headers = ["Product_Name", "Pricing", "Ratings"]
    writer.writerow(headers)

    # Step 4: Check if containers were found
    if containers:
        print(f"Number of containers found: {len(containers)}")

        # Step 5: Iterate through each container
        for container in containers:
            # Extract the title
            title_element = container.find("div", {"class": "s-item__title"})
            title = title_element.get_text(strip=True) if title_element else "Title not found"

            # Extract the price
            price_element = container.find("span", {"class": "s-item__price"})
            price = price_element.get_text(strip=True) if price_element else "Price not found"

            # Extract ratings if available
            rating_element = container.find("span", class_="s-item__reviews-count")
            ratings = rating_element.get_text(strip=True) if rating_element else "No ratings"

            # Write the row to the CSV file
            writer.writerow([title, price, ratings])

            # Optional: Print the results for each product
            print(f"Title: {title}, Price: {price}, Ratings: {ratings}")

    else:
        print("No containers found.")
```
